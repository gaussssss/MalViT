```python
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from src.data.tokenizer import load_config
config = load_config('../configs/config.yaml')
```

## Charger les images

```python
images_dir = Path('../data/images')
benign_imgs = sorted((images_dir / 'benign').glob('*.png'))
malware_imgs = sorted((images_dir / 'malware').glob('*.png'))
print(f'Benign: {len(benign_imgs)} | Malware: {len(malware_imgs)}')
```

## Grille 6x8 pour un fichier

```python
def show_grid(file_idx, label, images_dir, num_layers=6, num_heads=8):
    fig, axes = plt.subplots(num_layers, num_heads, figsize=(16,12))
    for l in range(num_layers):
        for h in range(num_heads):
            fname = f'file{file_idx:05d}_layer{l}_head{h}.png'
            p = images_dir / label / fname
            if p.exists():
                axes[l][h].imshow(np.array(Image.open(p)), cmap='hot')
            axes[l][h].axis('off')
        axes[l][0].set_ylabel(f'L{l}', fontsize=8)
    plt.suptitle(f'{label} - file {file_idx:05d}')
    plt.tight_layout()
    plt.show()
```

```python
show_grid(0, 'benign', images_dir)
show_grid(0, 'malware', images_dir)
```

## Comparaison benin vs malware - Layer 5, Head 0

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, label, cmap in zip(axes, ['benign','malware'], ['Blues','Reds']):
    p = images_dir / label / 'file00000_layer5_head0.png'
    if p.exists():
        ax.imshow(np.array(Image.open(p)), cmap=cmap)
        ax.set_title(f'{label} - Layer 5, Head 0')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
