# 01 - Tokenisation

```python
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from src.data.tokenizer import load_config, tokenize_file, compute_vocab_size
```

```python
config = load_config('../configs/config.yaml')
VOCAB_SIZE = compute_vocab_size(config['transformer']['token_bits'])
print(f'Vocabulary size: {VOCAB_SIZE}')
```

## Distribution des octets

```python
benign_files = list(Path('../data/raw/benign').glob('*.apk'))
malware_files = list(Path('../data/raw/malware').glob('*'))
print(f'Benign: {len(benign_files)}, Malware: {len(malware_files)}')
```

```python
if benign_files and malware_files:
    bt = tokenize_file(benign_files[0])
    mt = tokenize_file(malware_files[0])
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    axes[0].hist(bt, bins=256, color='steelblue', alpha=0.8)
    axes[0].set_title(f'Benign - {len(bt):,} tokens')
    axes[0].set_xlabel('Byte value')
    axes[0].set_ylabel('Frequency')
    axes[1].hist(mt, bins=256, color='crimson', alpha=0.8)
    axes[1].set_title(f'Malware - {len(mt):,} tokens')
    axes[1].set_xlabel('Byte value')
    plt.tight_layout()
    plt.show()
```

## Chunks

```python
from src.data.dataset import chunk_sequence
max_seq_len = config['transformer']['max_seq_len']
pad_token_id = config['transformer']['pad_token_id']
if benign_files:
    tokens = tokenize_file(benign_files[0])
    chunks = chunk_sequence(tokens, max_seq_len, pad_token_id)
    print(f'File: {len(tokens):,} bytes, {len(chunks)} chunks')
    print(f'PAD in last chunk: {np.sum(chunks[-1] == pad_token_id)}')
```

```python
if benign_files:
    plt.figure(figsize=(6,6))
    plt.imshow(chunks[0].reshape(32,32), cmap='gray', vmin=0, vmax=255)
    plt.title('First chunk - Benign (32x32)')
    plt.colorbar(label='Byte value')
    plt.show()
```
